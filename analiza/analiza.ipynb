```python
import pandas
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
```

```python
A = [join("A", f) for f in listdir("A") if isfile(join("A", f))]
```

```python
for a in A:
    data = pandas.read_csv(a, sep=" ", names=["time", "x", "y", "z"])
    
    plt.figure()
    data['x'].plot()
    data['y'].plot()
    data['z'].plot()
```
